```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Top countries GDP info.csv')
df
```

```python
df_clean = df.dropna()
df_clean
```

**U.S. Data**

```python
df_US = df_clean[df_clean['Country'] == 'US']
df_US
```

```python
df_US.plot(x='Year', y='GDP (in Bil. US$PPP)', kind='line', title='United States GDP from 2001 to 2027')
plt.show()
```

**China Data**

```python
df_China = df_clean[df_clean['Country'] == 'China']
df_China
```
